```python
import pandas as pd
import numpy as np
from itertools import product

# Hàm Haversine tính khoảng cách (km)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi/2)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(dlambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# -----------------------------
# 1. Tọa độ thủ đô
# -----------------------------
capitals = pd.DataFrame([
    ["Cambodia","Phnom Penh",11.5625,104.916],
    ["China","Beijing",39.9056,116.3958],
    ["Fiji","Suva",-18.1416,178.441],
    ["Indonesia","Jakarta",-6.1862,106.8063],
    ["Kiribati","South Tarawa",1.338,172.978],
    ["Lao People's Dem. Rep.","Vientiane",17.9689,102.6137],
    ["Malaysia","Kuala Lumpur",3.1502,101.7077],
    ["Mongolia","Ulaanbaatar",47.9138,106.922],
    ["Myanmar","Naypyidaw",19.754,96.1345],
    ["Palau","Koror",7.34,134.47],
    ["Papua New Guinea","Port Moresby",-9.46,147.19],
    ["Philippines","Manila",14.60,120.98],
    ["Samoa","Apia",-13.83,-171.73],
    ["Solomon Islands","Honiara",-9.43,159.95],
    ["Thailand","Bangkok",13.7573,100.502],
    ["Timor-Leste","Dili",-8.5662,125.588],
    ["Tonga","Nuku'alofa",-21.13,-175.20],
    ["Viet Nam","Hanoi",21.03,105.85],
    ["Australia","Canberra",-35.28,149.13],
    ["Brunei Darussalam","Bandar Seri Begawan",4.9431,114.9425],
    ["China, Hong Kong SAR","Hong Kong",22.3193,114.1694],
    ["China, Macao SAR","Macau",22.1987,113.5439],
    ["FS Micronesia","Palikir",6.9172,158.1851],
    ["Japan","Tokyo",35.6785,139.6823],
    ["Marshall Isds","Majuro",7.1004,171.38],
    ["New Zealand","Wellington",-41.2865,174.7762],
    ["Rep. of Korea","Seoul",37.5139,126.9828],
    ["Singapore","Singapore",1.2894,103.85],
    ["Tuvalu","Funafuti",-8.5243,179.1942],
    ["Vanuatu","Port Vila",-17.733,-168.33]
], columns=["country","capital","lat","lon"])

# -----------------------------
# 2. Tạo tất cả cặp kết hợp (origin-destination)
# -----------------------------
pairs = pd.merge(capitals, capitals, how='cross', suffixes=('_o','_d'))

# Tính khoảng cách
pairs['distance_km'] = pairs.apply(
    lambda row: haversine(row['lat_o'], row['lon_o'], row['lat_d'], row['lon_d']), axis=1
)

# Lọc chỉ lấy cột cần thiết
pairs = pairs[['country_o','capital_o','country_d','capital_d','distance_km']]

# Nếu muốn: loại bỏ trùng A-B = B-A và A-A
pairs = pairs[pairs['country_o'] <= pairs['country_d']].reset_index(drop=True)

# -----------------------------
# 3. Xuất ra CSV hoặc xem
# -----------------------------
pairs.to_csv("capital_distances.csv", index=False)
print(pairs.head(20))
```
