## For each country, add the value of its exports with Vietnam and the value of its imports from Vietnam to get its trade volume with Vietnam.

```python
# Exporters from file export_1
vietnam_exports1 = export1[export1['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_exports1)
```

```python
# Exporters from file export_2
vietnam_exports2 = export2[export2['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_exports2)
```

```python
# Exporters from file export_3
vietnam_exports3 = export3[export3['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_exports3)
```

```python
# Exporters from file import_1
vietnam_imports1 = import1[import1['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_imports1)
```

```python
# Exporters from file import_2
vietnam_imports2 = import2[import2['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_imports2)
```

```python
# Exporters from file import_3
vietnam_imports3 = import3[import3['partner_name'] == 'Viet Nam'][['reporter_name', 'partner_name', 'value']]
display(vietnam_imports3)
```

```python
# Combine and sum export values
combined_vietnam_exports = pd.concat([vietnam_exports1, vietnam_exports2, vietnam_exports3])
vietnam_export_summary = combined_vietnam_exports.groupby('reporter_name')['value'].sum().reset_index()
vietnam_export_summary = vietnam_export_summary.rename(columns={'value': 'export_value (VN)'})

# Combine and sum import values
combined_vietnam_imports = pd.concat([vietnam_imports1, vietnam_imports2, vietnam_imports3])
vietnam_import_summary = combined_vietnam_imports.groupby('reporter_name')['value'].sum().reset_index()
vietnam_import_summary = vietnam_import_summary.rename(columns={'value': 'import_value (VN)'})

# Merge export and import summaries
vietnam_trade_summary = pd.merge(vietnam_export_summary, vietnam_import_summary, on='reporter_name', how='outer').fillna(0)

# Calculate trade volume
vietnam_trade_summary['trade_volume (VN)'] = vietnam_trade_summary['export_value (VN)'] + vietnam_trade_summary['import_value (VN)']

# Display the desired columns
display(vietnam_trade_summary[['reporter_name', 'export_value (VN)', 'import_value (VN)', 'trade_volume (VN)']])
```

## Add up each country's trade volume with Vietnam to get Southeast Asia's trade volume with Vietnam.

```python
vietnam_trade_summary_2019 = vietnam_trade_summary['trade_volume (VN)'].sum()
print(f"Southeast Asia's trade volume with Vietnam is: {vietnam_trade_summary_2019}")
```

## Divide each country's trade volume with Vietnam by Southeast Asia's trade volume with Vietnam to obtain its share of Southeast Asia's trade with Vietnam.

```python
vietnam_trade_summary['Percentage Share'] = (vietnam_trade_summary['trade_volume (VN)']/vietnam_trade_summary_2019) * 100
display(vietnam_trade_summary)
```

```python
import altair as alt

# Sort the data by percentage share in descending order (still useful for order)
vietnam_trade_summary_sorted = vietnam_trade_summary.sort_values(by='Percentage Share', ascending=False)

# Create the base chart
base = alt.Chart(vietnam_trade_summary_sorted).encode(
    theta=alt.Theta("trade_volume (VN)", stack=True) # Removed sort="-theta" in previous fix
)

# Create the pie chart - Adding a subtle stroke for a slight 3D effect
pie = base.mark_arc(outerRadius=120, stroke="#ffffff", strokeWidth=1).encode(
    color=alt.Color(
        "Percentage Share",
        type="quantitative", # Encode a quantitative field
        scale=alt.Scale(range="heatmap"), # Use a sequential scheme where darker means higher values
        legend=alt.Legend(title="Trade Share (%)") # Add a legend for the color scale
    ),
    order=alt.Order("trade_volume (VN)", sort="descending"), # Order slices by size
    tooltip=[
        "reporter_name",
        alt.Tooltip("trade_volume (VN)", title="Trade Volume (USD)"),
        alt.Tooltip("Percentage Share", format=".2f", title="Percentage Share (%)")
    ]
)

# Add text labels (Country Name) for top 4 only
top_4_countries = vietnam_trade_summary_sorted.head(4)['reporter_name'].tolist()
text = base.mark_text(radius=140).encode(
    text=alt.Text("reporter_name"), # Display country name
    order=alt.Order("trade_volume (VN)", sort="descending"),
    color=alt.value("black"), # Text color
    opacity=alt.condition(
        alt.FieldOneOfPredicate(field='reporter_name', oneOf=top_4_countries),
        alt.value(1), # Show text for top 4
        alt.value(0)  # Hide text for others
    )
)


# Combine the pie chart and text labels and add title and font
chart = (pie + text).properties(
    title="Percentage Share of Trade with Vietnam (2019)"
).configure_text(
    font='lato'
)

# Display the chart
chart.display()
```
